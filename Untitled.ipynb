# Pre-Processing : Tokenization, Stemming, Lemmatization, Normalization

```python

def remove_punct(text):
    text = str(text)
    text = re.sub(r"(?i)\b((?:https?://|www\d{0,3}[.]|[a-z0-9.\-]+[.][a-z]{2,4}/)(?:[^\s()<>]+|\(([^\s()<>]+|(\([^\s()<>]+\)))*\))+(?:\(([^\s()<>]+|(\([^\s()<>]+\)))*\)|[^\s`!()\[\]{};:'\".,<>?«»“”‘’]))", '', text)
    text  = "".join([char for char in text if char not in string.punctuation])
    text = re.sub('[0-9]+', '', text)
    return text

def tokenization(text):
    text = re.split('\W+', text)
    return text

def remove_stopwords(text):
    text = [word for word in text if word not in stopwords.words('english')]
    return text
    
ps = nltk.PorterStemmer()
def stemming(text):
    text = [ps.stem(word) for word in text]
    return text


from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
def lemmatize(text):
    text = [lemmatizer.lemmatize(word, pos='v') for word in text]
    return text
  
```

```python
df['tweet_punct'] = df['title'].apply(lambda x: remove_punct(x))
df['tweet_tokenized'] = df['tweet_punct'].apply(lambda x: tokenization(x.lower()))
df['tweet_nonstop'] = df['tweet_tokenized'].apply(lambda x: remove_stopwords(x))
df['tweet_stemmed'] = df['tweet_nonstop'].apply(lambda x: stemming(x))
```

```python
def jointext(text):
    join_text = ' '.join(text)
    return join_text

df['tweet_joined'] = df['tweet_stemmed'].apply(lambda x: jointext(x))
df['tweet_joined']
```

# Individual First Level Models

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
from sklearn import metrics
import time
classifier = MultinomialNB(0.1)
previous_score=0
final_score = []
i= -1
start = time.time()
for alpha in np.arange(0,1,0.1):
    i=i+1
    sub_classifier=MultinomialNB(alpha=alpha)
    sub_classifier.fit(tfidf_train,Y_train)
    y_pred=sub_classifier.predict(tfidf_test)
    score = metrics.accuracy_score(Y_test, y_pred)
    final_score.append(score)
    if score>previous_score:
        classifier=sub_classifier
    print("Alpha: {}, Score : {}".format(alpha,score))
stop = time.time()
print(f"Training time: {stop - start}s")    
```
